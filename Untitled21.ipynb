```python
# ═══════════════════════════════════════════════════════════════════
# Prunable CNN on CIFAR-10  ── FINAL / MAX SOLUTION
# Merges "max solution" pattern with all working fixes
# ═══════════════════════════════════════════════════════════════════
# CHANGES vs previous version:
#   • gate_scores init: zeros → randn  (matches "max solution" spec)
#   • gate_loss in training loop now uses model.modules() style
#     (more general: works even if you add more PrunableLinear layers)
#   • compute_sparsity also uses model.modules() style
#   • LAMBDAS = [1e-2, 0.1, 1.0]  (kept — old 1e-5..1e-3 gave 0% sparsity)
#   • Both SimpleCNN and ResNet backbones included; pick via MODEL_CLASS
# ───────────────────────────────────────────────────────────────────

# ══ [1] Imports & Config ════════════════════════════════════════════
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

DEVICE      = torch.device("cuda" if torch.cuda.is_available() else "cpu")
BATCH_SIZE  = 128
EPOCHS      = 30
LR          = 1e-3
LAMBDAS     = [1e-2, 0.1, 1.0]   # 1e-5..1e-3 are too small → 0% sparsity
GATE_THRESH = 1e-2                # gate < 0.01 → treated as pruned
SEED        = 42

torch.manual_seed(SEED)

# ── GPU check ────────────────────────────────────────────────────────
if torch.cuda.is_available():
    print(f"✅ GPU detected : {torch.cuda.get_device_name(0)}")
    print(f"   VRAM         : {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
else:
    print("⚠️  WARNING: No GPU found — running on CPU (will be very slow).")
    print("   In Colab: Runtime → Change runtime type → T4 GPU → Save → Restart session")

print(f"Device: {DEVICE}")


# ══ [2] Data ════════════════════════════════════════════════════════
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465),
                         (0.2470, 0.2435, 0.2616)),
])
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465),
                         (0.2470, 0.2435, 0.2616)),
])

train_set = torchvision.datasets.CIFAR10(
    root='./data', train=True,  transform=train_transform, download=True)
test_set  = torchvision.datasets.CIFAR10(
    root='./data', train=False, transform=test_transform,  download=True)

PIN = torch.cuda.is_available()   # pin_memory only works with GPU
train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=2, pin_memory=PIN)
test_loader  = DataLoader(test_set,  batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=2, pin_memory=PIN)
print(f"Train batches: {len(train_loader)} | Test batches: {len(test_loader)}")


# ══ [3] PrunableLinear ══════════════════════════════════════════════
class PrunableLinear(nn.Linear):
    """
    Drop-in replacement for nn.Linear with learnable gate scores.

    forward:  out = F.linear(x,  weight * sigmoid(gate_scores),  bias)

    gate_scores shape = (out_features, in_features)  — same as weight.
    L1 penalty on sigmoid(gate_scores) drives gates → 0 → weight pruning.

    Init: gate_scores ~ N(0,1)  →  sigmoid ≈ 0.27–0.73 (random start).
    """
    def __init__(self, in_features: int, out_features: int):
        super().__init__(in_features, out_features)
        # randn init: gates start random in (0,1) — as per "max solution"
        self.gate_scores = nn.Parameter(
            torch.randn(out_features, in_features))

    def forward(self, x):
        gates         = torch.sigmoid(self.gate_scores)   # (out, in) ∈ (0,1)
        pruned_weight = self.weight * gates               # element-wise mask
        return F.linear(x, pruned_weight, self.bias)

    def sparsity(self, threshold=GATE_THRESH) -> float:
        """Fraction of this layer's gates below threshold."""
        with torch.no_grad():
            gates = torch.sigmoid(self.gate_scores)
            return (gates < threshold).float().mean().item()


# ══ [4a] Simple CNN backbone (no residual blocks) ═══════════════════
class SimpleCNNPrunable(nn.Module):
    """
    Plain CNN + PrunableLinear classifier head.
    Faster to train than ResNet; ideal for quick experiments.

    Conv 3→32  → BN → ReLU → MaxPool       [16×16]
    Conv 32→64 → BN → ReLU → MaxPool       [ 8×8]
    Conv 64→128→ BN → ReLU → AvgPool       [ 1×1]
    PrunableLinear(128 → 256) → ReLU
    PrunableLinear(256 → 128) → ReLU
    PrunableLinear(128 →  10) → logits
    """
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d( 3,  32, 3, padding=1, bias=False),
            nn.BatchNorm2d(32), nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                                   # 16×16

            nn.Conv2d(32,  64, 3, padding=1, bias=False),
            nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                                   #  8×8

            nn.Conv2d(64, 128, 3, padding=1, bias=False),
            nn.BatchNorm2d(128), nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d(1),                           #  1×1
        )
        self.fc1  = PrunableLinear(128, 256)
        self.fc2  = PrunableLinear(256, 128)
        self.fc3  = PrunableLinear(128,  10)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        x = self.features(x).flatten(1)
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return self.fc3(x)


# ══ [4b] ResNet-style backbone ══════════════════════════════════════
class ResBlock(nn.Module):
    def __init__(self, in_ch, out_ch, stride=1):
        super().__init__()
        self.conv1 = nn.Conv2d(in_ch, out_ch, 3,
                               stride=stride, padding=1, bias=False)
        self.bn1   = nn.BatchNorm2d(out_ch)
        self.conv2 = nn.Conv2d(out_ch, out_ch, 3,
                               stride=1, padding=1, bias=False)
        self.bn2   = nn.BatchNorm2d(out_ch)
        self.relu  = nn.ReLU(inplace=True)
        self.shortcut = nn.Sequential()
        if stride != 1 or in_ch != out_ch:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_ch, out_ch, 1, stride=stride, bias=False),
                nn.BatchNorm2d(out_ch),
            )

    def forward(self, x):
        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += self.shortcut(x)
        return self.relu(out)


class ResNetPrunable(nn.Module):
    """
    ResNet-style backbone + PrunableLinear head.

    Init conv   : 3 → 16  (32×32)
    Stage 1     : 16 → 16 (32×32) × 2 blocks
    Stage 2     : 16 → 32 (16×16) × 2 blocks, stride-2
    Stage 3     : 32 → 64 ( 8×8)  × 2 blocks, stride-2
    GAP         : 64-dim vector
    PrunableLinear(64 → 128) → ReLU
    PrunableLinear(128→  64) → ReLU
    PrunableLinear( 64→  10) → logits
    """
    def __init__(self):
        super().__init__()
        self.init_conv = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1, bias=False),
            nn.BatchNorm2d(16), nn.ReLU(inplace=True),
        )
        self.stage1 = self._make_stage(16, 16, n=2, stride=1)
        self.stage2 = self._make_stage(16, 32, n=2, stride=2)
        self.stage3 = self._make_stage(32, 64, n=2, stride=2)
        self.gap    = nn.AdaptiveAvgPool2d(1)
        self.fc1    = PrunableLinear(64, 128)
        self.fc2    = PrunableLinear(128, 64)
        self.fc3    = PrunableLinear(64,  10)
        self.relu   = nn.ReLU(inplace=True)

    def _make_stage(self, in_ch, out_ch, n, stride):
        layers = [ResBlock(in_ch, out_ch, stride)]
        for _ in range(n - 1):
            layers.append(ResBlock(out_ch, out_ch, stride=1))
        return nn.Sequential(*layers)

    def forward(self, x):
        x = self.init_conv(x)
        x = self.stage1(x)
        x = self.stage2(x)
        x = self.stage3(x)
        x = self.gap(x).flatten(1)
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return self.fc3(x)


# ── Pick backbone here ───────────────────────────────────────────────
MODEL_CLASS = SimpleCNNPrunable   # swap to ResNetPrunable if desired

_m = MODEL_CLASS()
print("Output shape :", _m(torch.randn(2, 3, 32, 32)).shape)   # [2, 10]
print(f"Total params : {sum(p.numel() for p in _m.parameters()):,}")
del _m


# ══ [5] Training utilities ══════════════════════════════════════════

def compute_gate_loss(model):
    """
    Gate-sparsity loss: Σ sigmoid(gate_scores) over ALL PrunableLinear layers.
    Uses model.modules() — automatically picks up any future layers added.

    WHY sigmoid first:
      gate_scores are unbounded reals; sigmoid maps them to (0,1).
      The L1 penalty on sigmoid values directly penalises "open" gates.
    """
    gate_loss = 0.0
    for module in model.modules():
        if isinstance(module, PrunableLinear):
            gates = torch.sigmoid(module.gate_scores)
            gate_loss = gate_loss + gates.sum()
    return gate_loss


def compute_sparsity(model, threshold=GATE_THRESH):
    """
    % of gates whose sigmoid value < threshold (treated as pruned).
    Uses model.modules() — same iteration style as compute_gate_loss.
    """
    num_zeros   = 0
    total_gates = 0
    with torch.no_grad():
        for module in model.modules():
            if isinstance(module, PrunableLinear):
                gates = torch.sigmoid(module.gate_scores).cpu().numpy()
                total_gates += gates.size
                num_zeros   += (gates < threshold).sum()
    return 100.0 * num_zeros / total_gates


def train_one_epoch(model, loader, optimizer, criterion, lam, device):
    """One epoch: CrossEntropy + λ × gate_loss."""
    model.train()
    total_loss = 0.0
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()

        logits    = model(imgs)
        ce_loss   = criterion(logits, labels)
        gate_loss = compute_gate_loss(model)          # Σ sigmoid(gate_scores)
        loss      = ce_loss + lam * gate_loss         # combined objective

        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)


def evaluate(model, loader, device):
    """Test accuracy (0–1 float)."""
    model.eval()
    correct = total = 0
    with torch.no_grad():
        for imgs, labels in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            preds    = model(imgs).argmax(dim=1)
            correct += (preds == labels).sum().item()
            total   += labels.size(0)
    return correct / total


def print_gate_stats(model):
    """Per-layer gate min / max / mean / near-zero %."""
    for module in model.modules():
        if isinstance(module, PrunableLinear):
            g = torch.sigmoid(module.gate_scores).detach().cpu()
            nz = (g < GATE_THRESH).float().mean() * 100
            print(f"    gate: min={g.min():.4f}  max={g.max():.4f}  "
                  f"mean={g.mean():.4f}  near-zero={nz:.1f}%")


# ══ [6] Experiment loop ══════════════════════════════════════════════
results          = []
criterion        = nn.CrossEntropyLoss()
best_model_state = None
best_acc         = 0.0

for lam in LAMBDAS:
    print(f"\n{'='*60}")
    print(f"  λ = {lam}   backbone: {MODEL_CLASS.__name__}")
    print(f"{'='*60}")

    torch.manual_seed(SEED)
    model     = MODEL_CLASS().to(DEVICE)
    optimizer = optim.Adam(model.parameters(), lr=LR, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

    for epoch in range(1, EPOCHS + 1):
        train_loss = train_one_epoch(
            model, train_loader, optimizer, criterion, lam, DEVICE)
        scheduler.step()

        if epoch % 5 == 0 or epoch == 1:
            val_acc = evaluate(model, test_loader, DEVICE)
            sp      = compute_sparsity(model)
            print(f"  Epoch {epoch:3d}/{EPOCHS} | "
                  f"loss {train_loss:.4f} | "
                  f"acc {val_acc*100:.2f}% | "
                  f"sparsity {sp:.1f}%")

    final_acc = evaluate(model, test_loader, DEVICE)
    final_sp  = compute_sparsity(model)
    results.append({"lambda": lam, "acc": final_acc * 100, "sparsity": final_sp})

    print(f"\n  → Final acc: {final_acc*100:.2f}%  |  Sparsity: {final_sp:.1f}%")
    print("  Gate stats (per PrunableLinear layer):")
    print_gate_stats(model)

    if final_acc > best_acc:
        best_acc         = final_acc
        best_model_state = model.state_dict()

print("\nAll experiments done.")


# ══ [7] Results table ═══════════════════════════════════════════════
print(f"\n{'─'*50}")
print(f"{'Lambda':>12} | {'Test Acc (%)':>12} | {'Sparsity (%)':>12}")
print(f"{'─'*50}")
for r in results:
    print(f"{r['lambda']:>12.0e} | {r['acc']:>12.2f} | {r['sparsity']:>12.1f}")
print(f"{'─'*50}")


# ══ [8] Gate histogram — best model ═════════════════════════════════
best_model = MODEL_CLASS().to(DEVICE)
best_model.load_state_dict(best_model_state)
best_model.eval()

layer_labels  = []
all_gate_vals = []
layer_colors  = ['#534AB7', '#1D9E75', '#D85A30']

for i, module in enumerate(
        m for m in best_model.modules() if isinstance(m, PrunableLinear)):
    g = torch.sigmoid(module.gate_scores).detach().cpu().numpy().ravel()
    all_gate_vals.append(g)
    label = f"fc{i+1} (n={len(g):,})"
    layer_labels.append(label)
    print(f"{label}  mean={g.mean():.4f}  "
          f"near-zero: {(g < GATE_THRESH).mean()*100:.1f}%")

all_gates_flat = np.concatenate(all_gate_vals)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: per-layer
for g, lbl, col in zip(all_gate_vals, layer_labels,
                        layer_colors[:len(all_gate_vals)]):
    axes[0].hist(g, bins=80, alpha=0.6, label=lbl, color=col, range=(0, 1))
axes[0].axvline(GATE_THRESH, color='red', linestyle='--',
                label=f'threshold ({GATE_THRESH})')
axes[0].set_title("Gate distribution (per layer)")
axes[0].set_xlabel("sigmoid(gate_score)")
axes[0].set_ylabel("count")
axes[0].legend(fontsize=8)

# Right: all combined
axes[1].hist(all_gates_flat, bins=100, color='#534AB7',
             alpha=0.8, range=(0, 1))
axes[1].axvline(GATE_THRESH, color='red', linestyle='--',
                label=f'threshold ({GATE_THRESH})')
axes[1].set_title("All gates combined (bimodal = healthy pruning)")
axes[1].set_xlabel("sigmoid(gate_score)")
axes[1].set_ylabel("count")
axes[1].legend(fontsize=8)

best_lam = results[np.argmax([r['acc'] for r in results])]['lambda']
plt.suptitle(
    f"Gate histogram — best model  λ={best_lam:.0e}  "
    f"acc={best_acc*100:.2f}%",
    fontsize=12, y=1.02)
plt.tight_layout()
plt.savefig("gate_histogram.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved → gate_histogram.png")


# ══ [9] Save checkpoint ══════════════════════════════════════════════
torch.save({
    "model_state_dict" : best_model_state,
    "results"          : results,
    "epochs"           : EPOCHS,
    "lambdas_tested"   : LAMBDAS,
    "model_class"      : MODEL_CLASS.__name__,
}, "best_model.pth")
print("Checkpoint saved → best_model.pth")
```
